```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

horspool = pd.read_csv("horspool_values_PES2UG23CS048.txt")
boyermoore = pd.read_csv("boyermoore_values_PES2UG23CS048.txt")

horspool.columns = ['patternlen', 'textlen', 'cmp', 'timetaken']
boyermoore.columns = ['patternlen', 'textlen', 'cmp', 'timetaken']

horspool['timetaken_ms'] = horspool['timetaken'] / 1e6
boyermoore['timetaken_ms'] = boyermoore['timetaken'] / 1e6

plt.figure(figsize=(10, 6))
plt.plot(horspool['textlen'], horspool['timetaken_ms'], label='Horspool', marker='o')
plt.plot(boyermoore['textlen'], boyermoore['timetaken_ms'], label='Boyer-Moore', marker='s')
plt.xlabel('Text Length')
plt.ylabel('Execution Time (ms)')
plt.title('Text Length vs Execution Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(horspool['textlen'], horspool['cmp'], label='Horspool', marker='o')
plt.plot(boyermoore['textlen'], boyermoore['cmp'], label='Boyer-Moore', marker='s')
plt.xlabel('Text Length')
plt.ylabel('Number of Comparisons')
plt.title('Text Length vs Number of Comparisons')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
